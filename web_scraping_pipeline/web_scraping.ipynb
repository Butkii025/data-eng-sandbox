```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://quotes.toscrape.com/page/1/"

res = requests.get(url)
if res.status_code == 200:
    print(res)
else:
    print("failed to fetch")
```

## from quotes, extract all quotes & authors from all pages with the "life" tag.

```python
#code
page_count=1

while True:
    url=f"https://quotes.toscrape.com/page/{page_count}/"
    res=requests.get(url)

    soup= BeautifulSoup(res.text,"lxml")
    quote = soup.select("div.quote")

    if not quote:
        print("no value page anymore...")
        break
    with open(f"scraped_data/quote{page_count}.html","w", encoding="utf-8") as f:
        f.write(res.text)
        print(f"downloaded data from page {page_count}")
    page_count= page_count+1
    
```

## extract insights

```python
# for page one

with open("scraped_data/quote2.html","r", encoding="utf-8") as f:
    content=f.read()
soup= BeautifulSoup(content,"lxml")
```

```python
all_quote=soup.select("div.quote")
life_quote=[]

# print(all_quote)

for q in all_quote:
    all_tags=[]
    for tag in q.select(".tags .tag"):
        all_tags.append(tag.get_text())

    if "life" in all_tags:
        text=q.select_one("span.text").get_text()
        # print(text)
        author=q.select_one("small.author").get_text()
        life_quote.append([text,author])
        
print(life_quote)
    
```

```python
import pandas as pd
df=pd.DataFrame(life_quote)
```

```python
df.to_json("life_quote_page1.json")
```
