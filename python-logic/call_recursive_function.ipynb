## Program to:- Implement recursive functions (factorial, Fibonacci)- Compare with iterative approach

```python
# Recursive factorial
def fact_rec(n):
    if n == 0 or n==1 :
        return 1
    return n * fact_rec(n-1)

# factorialIterative 
def fact_iter(n):
    res = 1
    for i in range(1, n+1):
        res *= i
    return res

# Fibonacci (Recursive)
def fib_rec(n):
    if n <= 1:
        return n
    return fib_rec(n - 1) + fib_rec(n - 2)

# Fibonacci (Iterative)
def fib_iter(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

#input
n = int(input("Enter a number: "))

# Output
print("\nFactorial:")
print("Recursive:", fact_rec(n))
print("Iterative:", fact_iter(n))

print("\nFibonacci:")
print("Recursive:", fib_rec(n))
print("Iterative:", fib_iter(n))
```
