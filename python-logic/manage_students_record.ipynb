## Program to manage student records using a dictionary:- Add student (name, marks)- Display all students- Find topper using function

```python
students = {}

def add_student(name, marks):
    students[name] = marks

def display():
    print(students)

def topper():
    top = max(students, key=students.get)
    print(f"Topper is {top} with marks {students[top]}")

add_student("A", 80)
add_student("B", 95)
display()
topper()
```
