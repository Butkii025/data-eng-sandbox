## program to count the frequency of each character in a string using a dictionary also display the most frequent character.

```python
string = input("Enter a string: ")

freq = {}
for char in string:
    freq[char] = freq.get(char, 0) + 1

print("Character Frequency:", freq)

max_freq = max(freq.values())

# Find all characters with that frequency
most_freq_chars = [char for char in freq if
                   freq[char] == max_freq]

print("Most frequent character:", most_freq_chars)
```
