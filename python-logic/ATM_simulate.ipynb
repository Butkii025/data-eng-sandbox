##  Program to simulate a bank account system with:
a. Deposit, withdraw, and balance check functions
b. Exception handling for insufficient balance
c. Menu-driven interface using loops

```python
#### code 

balance = 0

while True:
    print("\n1. Deposit\n2. Withdraw\n3. Check Balance\n4. Exit")
    choice = int(input("Enter choice: "))

    try:
        if choice == 1:
            amt = float(input("Enter amount: "))
            if amt <=0:
                raise Exception("enter sufficiant amount")

            balance += amt
            print("Deposited successfully")


        elif choice == 2:
            amt = float(input("Enter amount: "))
            if amt > balance and amt ==0:
                raise Exception("Insufficient Balance")
            balance -= amt
            print("Withdrawn successfully")

        elif choice == 3:
            print("Balance:", balance)

        elif choice == 4:
            break

        else:
            print("choice must be in range of 1-4")


    except Exception as e:
        print("Error:", e)
```
