## Program to handle multiple exceptions:-
ZeroDivisionError- ValueError

```python
try:
    a = int(input("Enter number: "))
    b = int(input("Enter number: "))
    print("Result:", a / b)


except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Invalid input")
```

## Program to demonstrate string manipulation:-
Reverse- Palindrome check- Slicing

```python
# code
s = input("Enter string: ")

# Reverse
print("Reverse:", s[::-1])

# Palindrome
if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")

# Slicing
print("First 3 chars:", s[:3])
```

## ques 5: Using NumPy and Pandas, write a program to:- 
Create a NumPy array- Convert it into a Pandas DataFrame- Display basic statistics (mean, max, min)

```python
import numpy as np
import pandas as pd

arr = np.array([10, 20, 30, 40])
print(f"array is:{arr}",type(arr))
df = pd.DataFrame(arr, columns=["Values"])

print(df, type(df))

print("Mean:", df["Values"].mean())
print("Max:", df["Values"].max())
print("Min:", df["Values"].min())
```
