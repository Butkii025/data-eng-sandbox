## Program using Matplotlib / Seaborn to:- Take a list of numbers- Plot a bar chart and histogram- Add labels and titl

```python
import matplotlib.pyplot as plt

nums = [10, 20, 30, 40]
x=[1,2,3,4]

# Bar chart
plt.bar(x,nums)
plt.title("Bar Chart")
plt.xlabel("Index")
plt.ylabel("Values")
plt.show()

# Histogram
plt.hist(nums)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()


# line plot

plt.plot(x,nums)
plt.title("line plot")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()


# plt.scatter(nums)
plt.scatter(x,nums)
plt.title("scatter plot")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
