```python
import pandas as pd
df=pd.read_csv("IRIS.csv")
```

```python
# questions

# first 10 rows
df.head(10)

# shape & data_type
df.shape  # no. of rows and columns
type(df) # pandas.core.frame

# summary satatics (mean, min, max, std)
df["sepal_length"].mean()
df["sepal_length"].min()
df["sepal_length"].max()
df["sepal_length"].std()

# petal_lenght > 4.5
df[df["petal_length"]>=4.5]

# species = "Iris-virginica"
df[df["species"]=="Iris-virginica"]

# groupby species & avg sepal_length
df.groupby("species")["sepal_length"].mean()


# max petal_length
df.groupby("species")["petal_width"].max()

# standard deviation of sepal_width
df.groupby("species")["sepal_width"].std()

# add new column of petal_ratio
df2=df.copy()
df2=df2.assign(petal_ratio=df["petal_length"]/df["petal_width"] )
df2
```

```python
df2.to_csv("Iris_filter.csv")
```
